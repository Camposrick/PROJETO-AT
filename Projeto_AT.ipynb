```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io

# Vamos buscar o nome do ficheiro que acabou de carregar
nome_do_ficheiro = list(uploaded.keys())[0]

# Lemos o ficheiro Excel
# skiprows=2 serve para ignorar as primeiras 2 linhas e apanhar o cabe√ßalho correto
df = pd.read_excel(io.BytesIO(uploaded[nome_do_ficheiro]), skiprows=2)

# O comando head() mostra apenas as primeiras 5 linhas para confirmarmos se est√° tudo bem
df.head()
```

```python
# 1. Filtramos apenas as linhas onde a Quantidade √© maior que 0
# Isso limpa as linhas vazias da planilha
producao_real = df[df['Quantidade'] > 0]

# 2. Fazemos a soma total de todos os r√°dios na planilha
total_geral = producao_real['Quantidade'].sum()

# 3. Fazemos a soma agrupada por 'Modelo'
# O 'groupby' junta tudo que tem o mesmo nome e o 'sum' soma as quantidades
soma_por_modelo = producao_real.groupby('Modelo')['Quantidade'].sum().reset_index()

print(f"‚úÖ Total de r√°dios produzidos pelo t√©cnico: {total_geral}")
print("\nüìä Quantidade por Modelo:")
print(soma_por_modelo)
```

```python
import pandas as pd
import os

# 1. Caminho da pasta onde voc√™ colocou os arquivos
pasta = 'producao'
todos_os_dados = []

# 2. Loop para ler cada arquivo na pasta
for arquivo in os.listdir(pasta):
    if arquivo.endswith('.xlsx'):
        caminho_completo = os.path.join(pasta, arquivo)

        # Lemos o arquivo (pulando as 2 linhas iniciais)
        # Adicionamos uma coluna com o nome do arquivo para saber de quem √© a produ√ß√£o
        df_temp = pd.read_excel(caminho_completo, skiprows=2)
        df_temp['Tecnico'] = arquivo

        todos_os_dados.append(df_temp)

# 3. Juntamos tudo em um √∫nico "Tabel√£o"
df_consolidado = pd.concat(todos_os_dados, ignore_index=True)

# 4. Limpeza: Manter apenas o que foi produzido
df_consolidado = df_consolidado[df_consolidado['Quantidade'] > 0]

# 5. Relat√≥rio Final: Soma por T√©cnico e por Modelo
relatorio = df_consolidado.groupby(['Tecnico', 'Modelo'])['Quantidade'].sum().reset_index()

print("üìä RELAT√ìRIO CONSOLIDADO GERADO!")
print(relatorio)
```

```python
# 1. Criamos o resumo agrupando por 'Status'
# Isso vai somar a 'Quantidade' para cada tipo de situa√ß√£o encontrada
resumo_status = df_consolidado.groupby('Status')['Quantidade'].sum().reset_index()

# 2. Exibimos o resultado na tela
print("üìã RELAT√ìRIO DE M√âTRICAS POR STATUS:")
print(resumo_status)

# 3. Para ver o total geral de TUDO o que foi processado:
total_geral = resumo_status['Quantidade'].sum()
print(f"\n‚úÖ Total Geral Processado: {total_geral} itens.")
```

```python
# 1. Soma total de r√°dios por t√©cnico (independente do modelo)
total_por_tecnico = df_consolidado.groupby('Tecnico')['Quantidade'].sum().reset_index()

# 2. Soma total de todos os r√°dios de todos os t√©cnicos juntos
total_geral_empresa = df_consolidado['Quantidade'].sum()

# --- Exibi√ß√£o dos resultados ---
print("üèÜ PRODU√á√ÉO TOTAL POR T√âCNICO:")
print(total_por_tecnico)

print("\n" + "="*40)
print(f"üöÄ TOTAL GERAL DE PRODU√á√ÉO (TODOS): {total_geral_empresa} r√°dios")
print("="*40)
```

```python
# 1. Converter a coluna 'Data' para o formato que o Python entende (Dia/M√™s/Ano)
df_consolidado['Data'] = pd.to_datetime(df_consolidado['Data'])

# 2. Criar uma nova coluna chamada 'Mes' extraindo apenas o M√™s e o Ano
# O c√≥digo '%m/%Y' transforma a data em algo como '01/2026'
df_consolidado['Mes'] = df_consolidado['Data'].dt.strftime('%m/%Y')

# 3. Vamos espreitar para ver se a nova coluna apareceu corretamente
print(df_consolidado[['Data', 'Mes', 'Status']].head())
```

```python
# 1. Agrupamos por 'Mes' e 'Status' ao mesmo tempo
# Assim ele separa cada m√™s e, dentro do m√™s, cada tipo de servi√ßo
relatorio_mensal = df_consolidado.groupby(['Mes', 'Status'])['Quantidade'].sum().reset_index()

# 2. Ordenamos para os meses aparecerem na ordem certa (opcional, mas ajuda a ler)
relatorio_mensal = relatorio_mensal.sort_values('Mes')

print("üìÖ RELAT√ìRIO FINAL DETALHADO POR M√äS:")
print(relatorio_mensal)
```
